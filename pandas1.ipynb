```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h1>My First Pandas Tutorial</h1>
<h2>importing CSV</h2>
<p>CSV is imported with the following code:</p>

```python
url = 'https://raw.githubusercontent.com/Rally0078/csv-repo/13f54c5fe47315c1a3c537e6231f7413a8a11f18/CPUtemps.CSV?raw=true'
cpu_temps = pd.read_csv(url, usecols = [u'CPU Package [°C]'], encoding = 'cp1252')
```

<h2>Displaying the DataFrame</h2>
<p>The DataFrame is displayed</p>

```python
cpu_temps
```

<h2>Cleaning the DataFrame to make it plottable:</h2>

```python
cpu_temps = cpu_temps.drop([241,242])
cpu_temps[u'CPU Package [°C]'] =cpu_temps[u'CPU Package [°C]'].astype(int)
cpu_temps[u'CPU Package [°C]'] =cpu_temps[u'CPU Package [°C]'].astype(int)
cpu_temps
```

<h2>Displaying the data type:</h2>
<p>The data type of the DataFrame object is displayed using DataFrame.dtypes</p>
<code>cpu_temps.dtypes</code>

```python
print(cpu_temps.dtypes)
```

<h2>Plotting the graph of temperature vs samples</h2>
<p>A graph of temperature is plotted against the number of sample, and the mean value of temperature is indicated with a dashed red line</p>

```python
cpu_temp_array = cpu_temps.to_numpy()
%matplotlib notebook
%matplotlib notebook
ax = cpu_temps.plot()
plt.axhline(y = np.mean(cpu_temp_array), color = 'r', linestyle = '--', label = u'Mean Temperature °C')
plt.xlabel('Samples')
plt.ylabel(u'Temperature °C')
plt.legend(bbox_to_anchor = (0.81, 1.15), loc = 'upper center')
```

<h2>Histogram of temperature:</h2>
<p>A histogram of the temperature is also plotted</p>

```python
bx = cpu_temps.plot.hist(bins = 12)
plt.xlabel(u'Temperature °C')
```
